```python
def without_ac(image):
    #requires ee, numpy as np
#    eq1 = a*ML+AL-oi


    ML = 0.00033420001273043454
    AL = 0.10000000149011612
    oi = 0.29
    ToA = image.expression(
              'im*ML+AL-oi',
                 {'im': image.select('B10'),
                  'ML': ML,
                  'AL': AL,
                  'oi': oi,
                 }).rename('ToA')
    
    
    #     eq to BT
    #     BT = k2/(np.log(((k1/eq1)+1)))-273.5
    k1 = 774.8853
    k2 = 1321.0789

    denominator = ToA.expression(
          '(k1/ToA)+1',
             {'k1': k1,
              'ToA':ToA,
             }).rename('band')




    # #ee log func
    denominator = denominator.log()
    
    BT = denominator.expression(
          '(k2/denom)-273.5',
             {'k2': k2,
              'denom':denominator,
             }).rename('BT')

    #    (BT/(1+((( 0.0010895*BT)/1.4388)*np.log(0.991))))

    emissivity = 0.991
    le = np.log(emissivity)
    
    WST = BT.expression(
          'BT/(1+(((0.0010895*BT)/1.4388)*le))',
        {'BT': BT, 
        'le':le}).rename('wst')

    return WST
```
